# Problem 1

Şimdi beraber kullanıcılardan aldığımız verileri işleyerek, onlara hem çıktı verecek, hem de aldığımız ve hesapladığımız verileri ileride kullanabilmek adına kaydedeceğiz.


 **Bu bir veri giriş platformu olacak**


 * kullanıcıya kaç giriş yapacağını soracağız
* kullanıcıdan isim-soyisim, yaş, boy, kilo bilgilerini alarak VKİ hesaplayacağız
* ardından bunları sınıflara göre kaydedeceğiz bir dosyaya kaydedeceğiz


Sınıflar ise şu şekilde: 
* 18, 5 kg/m.'nin altında olanlar: Zayıf
* 18.5 – 24, 9 kg/m. arasında olanlar: Normal kilolu
* 25 – 29, 9 kg/m. arasında olanlar: Fazla kilolu
* 30 – 39, 9 kg/m. arasında olanlar: Obez
* 40 kg/m.'nin üzerinde olanlar: İleri derecede obez (morbid obez), olarak görülür.

```python
# 1.Problemin Çözümünü Buraya Yazınız
import time

def vki_hesaplama(kilo,boy):
    vki= kilo/(boy**2)
    return round(vki,1)

def deger_araligi(vki):
    if vki< 18.5:
        return "zayıf"
    elif vki <= 24.9:
        return "normal kilo"
    elif vki <= 29.9:
        return "Fazla kilo"
    elif vki <=39.9:
        return "Obez"
    elif vki <=40:
        return "Ileri obez"
    
def cikti_olustur(isim,yas,kilo,boy,vki,deger):
        kayittarihi =time.ctime(time.time())

        t1= f"Kullanıcı adı {isim}\n"
        t2= f"Kullanıcını yas,kilo,boy bilgisi {yas},{kilo},{boy}\n"
        t2= f".Yas {yas}\n,Kilo {kilo}\n,Boy {boy}\n"
        t3= f"Vucut kitle endeksi{vki}\n"
        t4= f"Değer aralığı {deger}\n"
        t5= f"oluşturma tarihi {kayittarihi}"
        output = t1+t2+t3+t4+t5
        return output

def main():
     girissayisi= int(input("Kaç kişinin verisini gireceksiniz"))
     for i in range (girissayisi):
        print(f"\n{i+1}. kişi için bilgileri giriniz:")
        isim = input("Kullanıcı ismini giriniz:")
        yas = input("Kullanıcı yaşını giriniz:")
        kilo = float(input("Kullanıcı kilosunu giriniz:"))
        boy = float(input("Kullanıcı boyunu giriniz:"))

        vki = vki_hesaplama(kilo,boy)
        deger = deger_araligi(vki)

        with open("kayitlar.txt","a",encoding="utf-8")as dosya:
            dosya.write(cikti_olustur(isim,yas,kilo,boy,vki,deger))
            print(f"\n{i+1}. kişi kaydedildi.\n")
            print('-'*10)
main()


with open("kayitlar.txt","r",encoding="utf-8")as file:
    for i in file:
        print(i)
```

**YetBar - Smoothie Bank**

YetBar spor salonunun girişinde bulunan, insansız bir sağlıklı içecek otomatıdır. Kullanıcı çeşitli girdiler smoothie içeceğine koyacağı ürünleri belirleyebilir, fiyat hesaplaması yaptırabilir ya da sadece menüyü görüntüleyebilir.

**Projenin Amacı:**

*   Birbirini etkileyen ve bir kod içerisinde birden fazla defa başvurulacak fonksiyonların oluşturulup, "fonksiyonel" proglamlamak.
*   Büyük programları yazarken, yazılımcı olarak yapabileceğimiz hatalardan kaçınmak için kontrol yöntemleri geliştirmek.
*   Kullanıcı inputlarını ve yaptığımız hesapları, raporlamak/loglamak amacıyla kaydetmek.

**Programdan Beklenen Çıktılar**
* Program kullanıcıya temelde 2 farklı işlem yapma imkanı sağlar.
> 1. Menüyü görüntüleme
> 3. Satın alma
* Satın alınan her içecek tarih, fiyat bilgisi ile kaydedilmeli ve her zaman ulaşılabilir olmalıdır.




```python
# 2.Problemin Çözümünü Buraya Yazınız
import time

menu = {
    "elma" : 0.50,
    "armut" : 0.75,
    "karpuz" : 0.30,
    "çılek" : 0.50,
    "ananas" : 0.90,
    "kiraz" : 0.40
    }

def menugoster():
    print("************** YetGen Smoothie Menu**************\n")
    for item in list(menu.keys()):
         print(f"{item.upper():<10}, {menu[item]} TL'dir")



def tekurun_hesaplama(urun,adet):
    assert  adet >= 0 , "adet 0 ve 0'dan büyük yada küçük olamaz"

    if urun not in list(menu.keys()):
        return None
    else:
        return menu[urun]* adet
    
def satin_alma():

    menugoster()
    print("LÜtfen satın almak istediğiniz ürünleri giriniz")
    
    satinalinanurun = []
    satinalmaadeti = []

    son_fiyat = 0

    while 1:
        eklenecek_urunler = input("Eklemek istediğınız ürünü seçiniz")
        if eklenecek_urunler == "0":
            break

        eklemeadedi = int(input("Kaç adet eklemek istersiniz"))
        
        fiyat = tekurun_hesaplama(eklenecek_urunler, eklemeadedi)

        if fiyat is None:
            print("Yanlış bir işlem yaptınız. Ürün eklenemedi")

        else:
         print(f"Ürününüzün fiyatı {fiyat}. İşlemi onaylamak istiyorsanız 'y'. İptal etmek için 'n' giriniz")

         Kontrol = input("Satın almayı onaylıyormusunuz?")

         if Kontrol == "y":
             print("Satın alın onaylandı...")
             print('-'*40)
             son_fiyat+= fiyat
         elif Kontrol =="n":
             print("Satın alın iptal edildi...")
             print('-'*40)
    return son_fiyat, satinalmaadeti, satinalinanurun

def Kayit():
    print("YetGen Bara hoşgeldiniz!\n Menüyü görüntülemek için 1'e basınız\n smoothie yapmak için lütfen '2'e basınız\n Çıkmak için '3'e basınız.")
    Program = None

    while 1:
     Program = int(input("Tercihiniz"))

     if Program  <0 or Program >3:
        print("Yanliş işlem yaptınız. Tekrar deneyiniz")
     elif  Program == 3:
        print("Çıkış yapılıyor... \n Teşekkür ederiz :))")
        break
     elif Program == 1:
         menugoster()
     elif Program == 2:
        son_fiyat, satinalmaadeti, satinalinanurun = satin_alma() 
        with open("kasa.txt","a") as f:
             f.write(str(time.ctime(time.time())))
             f.write("\n")
             f.write(','.join(map(str, satinalinanurun)))
             f.write("\n")   
             f.write(','.join(map(str, satinalmaadeti))) 
             f.write("\n")
             f.write(f"Toplam: {son_fiyat}")
             f.write("\n")
             f.write("*"*20)
             f.write("\n")
         
         
             cikis_kontrol = input("Başka bir işlem yapmak istiyormusunuz (evet-hayır):")

             if cikis_kontrol == "evet":
               pass
             
             else:
               print("çıkış yapılıyor...\n TEŞEKKÜR EDERİZ")
               break
Kayit()
```
